```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from scipy.optimize import minimize
import matplotlib.pyplot as plt




pricing = pd.DataFrame({
    'prices': [100, 90, 70, 60, 40, 20, 10],
    'demand': [100, 120, 200, 300, 1200, 2000, 2800]
})

def revenue_maximizing_price(prices, demand):
    # Calculate revenue for each price level
    pricing['revenue'] = pricing['prices'] * pricing['demand']

    # Retrieve the optimal price and corresponding revenue using idxmax()
    optimal_price, max_revenue = pricing.loc[pricing['revenue'].idxmax(), ['prices', 'revenue']]

    return optimal_price, max_revenue

optimal_price, max_revenue = revenue_maximizing_price(pricing['prices'], pricing['demand'])

print("The optimal price that maximizes revenue is:", optimal_price)
print("The maximum revenue associated with the optimal price is:", max_revenue)

# Plotting the revenue curve
plt.plot(pricing['prices'], pricing['revenue'], label='Revenue Curve')

# Highlight the optimal price point
plt.scatter(optimal_price, max_revenue, color='red', label='Optimal Price Point')

# Adding labels and title
plt.xlabel('Prices')
plt.ylabel('Revenue')
plt.title('Revenue Maximization')
plt.legend()

# Display the plot
plt.show()
```
